```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
d1=sns.load_dataset('tips')
d1
```

```python
sns.lmplot(x='total_bill',y='tip',data=d1,fit_reg=False,height=6,col='day',row='sex',palette='summer')
plt.savefig('graph.jpg')
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',data=d1,jitter=False)
plt.show()
```

```python
sns.swarmplot(x='sex',y='tip',data=d1,hue='smoker')
plt.show()
```
