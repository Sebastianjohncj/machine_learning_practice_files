```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import statsmodels.api as sm
import scipy.stats as stats
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix #Evaluation of model
from sklearn.metrics import roc_curve,roc_auc_score
```

```python
train=pd.read_csv("Titanictrain.csv")
test=pd.read_csv("Titanictest.csv")
```

```python
train.head()
```

```python
test.head()
```

```python
train=train.drop("PassengerId",axis=1)
train=train.drop("Name",axis=1)
train=train.drop("Ticket",axis=1)
train=train.drop("Cabin",axis=1)
```

```python
train.head()
```

```python
ID=test.PassengerId
ID
```

```python
test = test.drop("Name", axis = 1)
test = test.drop("Ticket", axis = 1)
test = test.drop("Cabin", axis = 1)
test = test.drop("PassengerId", axis = 1)
```

```python
test.head()
```

```python
newtrain=pd.get_dummies(train)
```

```python
newtrain.head()
```

```python
newtest=pd.get_dummies(test)
```

```python
newtest.head()
```

```python
x=newtrain.drop("Survived",axis=1)
y=newtrain.Survived
```

```python
lg=LogisticRegression()
```

```python
lg.fit(x,y)
```

```python
pred_titanic=lg.predict(newtest)
```

```python
pred_titanic
```

```python
accuracy_score(y,lg.predict(x))
```

```python
newtest['Survived']=pred_titanic
```
