```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
%matplotlib inline #magic functions are the functions which have % before them
```

# line graph

```python
l1=list([2,3,4,5])
l1
```

```python
l1=[2,3,4,5]
l2=[3,4,5,2]
l3=[3,4,5,np.nan]
l4=[2,3,4,5]
```

```python
plt.plot(l1,l2)
plt.plot(l3,l4)
plt.show()
```

```python
plt.plot(l1,l2,label='firstline')#label for naming the line 
plt.plot(l3,l4,label='secondline')
plt.legend() #for displaying the legends
plt.title('linegraph') #for displaying the title
plt.xlabel('xaxis') #for displaying the xaxis name
plt.ylabel('yaxis')
plt.show() #for showing the graph clearly
```

```python
plt.plot(l1,l2,color='g',linewidth=3,linestyle='-')
plt.plot(l3,l4,color='r',linewidth=6,linestyle='-.')

plt.title('linegraph') #for displaying the title
plt.xlabel('xaxis') #for displaying the xaxis name
plt.ylabel('yaxis')
plt.grid() #will show the grids for the graph
plt.show()
```

```python
l1=[2,3,4,5]
l2=[3,4,5,2]
l3=[3,6,7,8]
l4=[1,7,6,9]
plt.plot(l1,l2,color='g',linewidth=3,linestyle='-',marker='o',label='price')
plt.plot(l3,l4,color='r',linewidth=6,linestyle='-.',marker='o',markersize='10',label='cost')
plt.legend()
plt.plot(l2,l4,'b.-.',marker='o',markersize='10')
plt.title('linegraph') #for displaying the title
plt.xlabel('xaxis') #for displaying the xaxis name
plt.ylabel('yaxis')
plt.grid() #will show the grids for the graph
plt.show()
```

```python
plt.plot(l1,l2,marker='s',markersize='6')
plt.plot(l3,l4,marker='*',markersize='10')
plt.plot(l4,l3,'r-.o')
plt.show()
```

```python
l1=np.arange(0,10,0.1)
s=np.sin(l1)
c=np.cos(l1)
s
c
```

```python
plt.plot(s,label='sine-wave')
plt.plot(c,label='cos-wave')
plt.legend()
plt.show()
```

```python
plt.plot(c)
plt.show()
```

```python
l1=np.arange(0,360,0.1)
s=np.sin(l1)
c=np.cos(l1)
s
c
plt.plot(s,c)
plt.show()
```

```python
l1=np.arange(0,10,0.1)
s=np.sin(l1)
c=np.cos(l1)
t=np.tan(l1)
s
c
t
#plt.plot(s,label='sine-wave')
#plt.plot(c,label='cos-wave')
plt.plot(t,label='tan-wave')
plt.legend()
plt.show()
```

# Bar Graph

```python
x=[3,5,7,9]
y=[12,15,16,8]
plt.bar(x,y,label='data')
plt.xlabel('xaxis')
plt.ylabel('yaxis')
plt.title('Bar graph')
plt.legend()
plt.show()
```

```python
x2=[2,4,6,8]
y2=[13,16,6,9]
plt.bar(x,y,label='data1',color='red')
plt.bar(x2,y2,label='data2',color='black')
plt.xlabel('xaxis')
plt.ylabel('yaxis')
plt.title('Bar graph')
plt.legend()
plt.show()
```

```python
x2=[2,4,6,8]
y2=[13,16,6,9]
plt.barh(x,y,label='data1',color='skyblue')
plt.barh(x2,y2,label='data2',color='k')
plt.xlabel('xaxis')
plt.ylabel('yaxis')
plt.title('Bar graph')
plt.legend()
plt.show()
```

```python
x2=[2,4,6,8]
y2=[13,16,6,9]
plt.bar(x,y,label='data1',color='skyblue')
plt.barh(x2,y2,label='data2',color='k')
plt.xlabel('xaxis')
plt.ylabel('yaxis')
plt.title('Bar graph')
plt.legend()
plt.show()
```

```python
df1=DataFrame({'company':['infosys','tcs','google','facebook','wipro','yahoo','amazon','flipkart'],
               'price':[100,200,900,500,200,50,1000,800]})
df1
```

```python
plt.bar(df1['company'],df1['price'])
plt.show()
```

```python
plt.barh(df1['company'],df1['price'], color='b',label='Stock price')
plt.legend()
plt.xlabel('Stock Price in USD')
plt.ylabel('Company Name')
plt.title('Stock price chart')
plt.show()
```
